-------------------------------------
- Suppose you start up a company that has developed a drug that is supposed to increase IQ. You know that the standard deviation of IQ in the general population is 15. You test your drug on 36 patients and obtain a mean IQ of 97.65. Using an alpha value of 0.05, is this IQ significantly different than the population mean of 100?

--------------------------------

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Given data
population_mean = 100
sample_mean     = 97.65
population_std  = 15
sample_size     = 36
alpha = 0.05
```

#### State the Null and Alternate Hypothesis

- Null Hypothesis (H0): The mean IQ of the sample is equal to the population mean.
    - ùêª0:ùúá=100

- Alternate Hypothesis (H1): The mean IQ of the sample is not equal to the population mean.
    - ùêª1:ùúá‚â†100


```python
# Calculate the Z-test statistic
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
z_stat
```

```python
# Critical values for two-tailed test at alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha/2)

# Calculate the p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))
```

```python
# Print the results
print(f"Z-test Statistic: {z_stat:.2f}")
print(f"Critical Values: ¬±{z_critical:.2f}")
print(f"P-value: {p_value:.4f}")
```

```python
# Plot the results
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)
```

```python
plt.plot(x, y, label='Standard Normal Distribution')

# Shade the rejection regions
plt.fill_between(x, 0, y, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.5, label='Rejection Region')

# Plot the test statistic
plt.axvline(z_stat, color='blue', linestyle='--', label=f'Test Statistic (Z = {z_stat:.2f})')

# Add labels and legend
plt.xlabel('Z-value')
plt.ylabel('Probability Density')
plt.title('Two-Tailed One-Sample Z-Test')
plt.legend()

# Show the plot
plt.show()
```

```python
# Conclusion
if abs(z_stat) > z_critical:
    print("Reject the null hypothesis: The mean IQ is significantly different from the population mean of 100.")
else:
    print("Fail to reject the null hypothesis: There is not enough evidence to suggest that the mean IQ is significantly different from the population mean of 100.")
```
