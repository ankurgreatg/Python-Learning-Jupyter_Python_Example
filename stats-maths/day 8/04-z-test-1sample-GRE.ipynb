```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Given data
population_mean = 210
sample_mean     = 212.79
population_std  = 8.5
sample_size     = 42
alpha = 0.05
```

```python
# Calculate the Z-test statistic
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
z_stat
```

```python
# Critical value for one-tailed test at alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha)
z_critical
```

```python
# Calculate the p-value
p_value = 1 - stats.norm.cdf(z_stat)
p_value
```

```python
# Print the results
print(f"Z-test Statistic: {z_stat:.2f}")
print(f"Critical Value: {z_critical:.2f}")
print(f"P-value: {p_value:.4f}")
```

```python
# Plot the results
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

plt.plot(x, y, label='Standard Normal Distribution')

# Shade the rejection region
plt.fill_between(x, 0, y, where=(x >= z_critical), color='red', alpha=0.5, label='Rejection Region')

# Plot the test statistic
plt.axvline(z_stat, color='blue', linestyle='--', label=f'Test Statistic (Z = {z_stat:.2f})')

# Add labels and legend
plt.xlabel('Z-value')
plt.ylabel('Probability Density')
plt.title('One-Sample Z-Test')
plt.legend()

# Show the plot
plt.show()
```

```python
# Conclusion
if z_stat > z_critical:
    print("Reject the null hypothesis: The new mean verbal GRE score is significantly greater than the desired mean of 210.")
else:
    print("Fail to reject the null hypothesis: There is not enough evidence to suggest that the new mean verbal GRE score is significantly greater than the desired mean of 210.")
```
