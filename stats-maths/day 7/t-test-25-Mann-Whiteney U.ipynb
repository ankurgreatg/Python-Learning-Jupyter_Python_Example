```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample data
group1 = [85, 86, 88, 75, 78]
group2 = [92, 94, 89, 95, 90]
```

```python
# Perform Mann-Whitney U Test
statistic, p_value = stats.mannwhitneyu(group1, group2)

print(f'Statistic: {statistic}')
print(f'p-value: {p_value}')
```

```python
# Define sample sizes
n1 = 5
n2 = 5
```

```python
# Calculate the mean and standard deviation of the U distribution
mean_U = n1 * n2 / 2
std_U  = np.sqrt(n1 * n2 * (n1 + n2 + 1) / 12)
```

```python
# Define the range of U values
U_values = np.arange(0, n1 * n2 + 1)
```

```python
# Calculate the pdf and cdf for the Mann-Whitney U distribution
pdf_U = stats.norm.pdf(U_values, mean_U, std_U)
cdf_U = stats.norm.cdf(U_values, mean_U, std_U)
```

```python
# Calculate the critical value for the 95% significance level
alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha, mean_U, std_U)
```

```python
# U statistic from the example calculation
U_statistic = 0
```

```python
# Plot the distribution
plt.figure(figsize=(12, 6))

# Plot PDF
plt.subplot(1, 2, 1)
plt.plot(U_values, pdf_U, label='PDF', color='blue')
plt.axvline(U_statistic, color='red', linestyle='dashed', linewidth=1, label=f'U statistic = {U_statistic}')
plt.axvline(critical_value, color='green', linestyle='dashed', linewidth=1, label=f'Critical value (95%) = {critical_value:.2f}')
plt.title('Mann-Whitney U Distribution - PDF')
plt.xlabel('U value')
plt.ylabel('Probability Density')
plt.legend()

# Plot CDF
plt.subplot(1, 2, 2)
plt.plot(U_values, cdf_U, label='CDF', color='blue')
plt.axvline(U_statistic, color='red', linestyle='dashed', linewidth=1, label=f'U statistic = {U_statistic}')
plt.axvline(critical_value, color='green', linestyle='dashed', linewidth=1, label=f'Critical value (95%) = {critical_value:.2f}')
plt.title('Mann-Whitney U Distribution - CDF')
plt.xlabel('U value')
plt.ylabel('Cumulative Probability')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Print the PDF and CDF values at U statistic
pdf_value = stats.norm.pdf(U_statistic, mean_U, std_U)
cdf_value = stats.norm.cdf(U_statistic, mean_U, std_U)

print(f'PDF value at U = {U_statistic}: {pdf_value}')
print(f'CDF value at U = {U_statistic}: {cdf_value}')
```
