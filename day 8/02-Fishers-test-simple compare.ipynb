Say we spend a few days counting whales and sharks in the Atlantic and Indian oceans. In the Atlantic ocean we find 8 whales and 1 shark, in the Indian ocean 2 whales and 5 sharks.

```python
from scipy.stats import fisher_exact
```

```python
# Create the contingency table
contingency_table = [[8, 2], 
                     [1, 5]]
```

```python
# Compute Fisher's exact test
odds_ratio, p_value = fisher_exact(contingency_table)
```

```python
print("p-value (scipy):", p_value)
```

1. Hypergeometric Distribution:

The hypergeometric distribution is a probability distribution that describes the probability of obtaining a specific number of successes (or items of interest) in a fixed number of draws from a finite population without replacement. It's used when the outcome of each draw is not independent of the previous draws because items are not replaced after they're drawn.

##### Terms in the Hypergeometric Probability Formula:

The hypergeometric probability formula involves several terms:

- N is the population size: It represents the total number of items in the population. In this context, it could be the total number of animals or observations.

- K is the number of successes in the population: It represents the total number of items of interest in the population. In this context, it could be the total number of whales or sharks.

- n is the sample size: It represents the number of draws or samples taken from the population. In this context, it could be the total number of animals observed in a sample.

- k is the number of successes in the sample: It represents the number of items of interest observed in the sample. In this context, it could be the number of whales or sharks observed in the sample.
