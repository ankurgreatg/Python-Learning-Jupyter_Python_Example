```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial
```

```python
def poisson_pmf(lambda_val, k):
    """
    Calculate the probability mass function (PMF) of the Poisson distribution.
    
    Parameters:
    lambda_val (float): The average rate of events occurring per unit of time or space.
    k (int or array-like): The number of events.
    
    Returns:
    float or array-like: The probability of observing k events.
    """
    return (lambda_val ** k) * np.exp(-lambda_val) / factorial(k)
```

```python
# just for testing
lambda_val = 10
k          = 10

poisson_pmf(lambda_val, k)
```

- The result 0.12 represents the probability associated with observing exactly 10 events (customer arrivals) in the specified scenario. 
- It indicates that there is approximately a 12% chance of exactly 10 customers arriving at the retail checkout counter within the given time interval, based on the Poisson distribution with an average arrival rate of 10 customers per hour.

```python
lambda_val = 10
k          = 9

poisson_pmf(lambda_val, k)
```

- This probability value indicates that there is approximately a 3.47% chance of observing exactly 15 customer arrivals at the retail checkout counter within the given time interval, based on the Poisson distribution with an average arrival rate of 10 customers per hour.

```python
# Generate values for k (number of events)
k_values = np.arange(0, 100)  # Considering up to 30 events

# Calculate the PMF for the given lambda value and k values
pmf_values = poisson_pmf(lambda_val, k_values)

# Plot the PMF
plt.figure(figsize=(10, 6))
plt.bar(k_values, pmf_values, color='skyblue')
plt.title('Poisson PMF for Arrival Rate of {:.2f} Customers per Hour'.format(lambda_val))
plt.xlabel('Number of Arrivals (k)')
plt.ylabel('Probability')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(k_values)
plt.show()
```

#### using scipy

```python
from scipy.stats import poisson
```

```python
# Example parameters
lambda_val = 10  # Average rate (lambda)
k          = 20  # Number of events
```

```python
# Calculate PMF (Probability Mass Function)
pmf_value = poisson.pmf(k, lambda_val)
print(f"The PMF for Poisson distribution with lambda={lambda_val} and k={k} is approximately {pmf_value:.4f}")

# Calculate CDF (Cumulative Distribution Function)
cdf_value = poisson.cdf(k, lambda_val)
print(f"The CDF for Poisson distribution with lambda={lambda_val} and k={k} is approximately {cdf_value:.4f}")
```
